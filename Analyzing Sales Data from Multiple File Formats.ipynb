```python
import pandas as pd
import numpy as np
import json

# Sample data for CSV format
csv_data = {
    'OrderID': [1, 2, 3, 4, 5],
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B'],
    'Sales': [100, 150, 230, 180, 200],
    'Region': ['East', 'West', 'North', 'South', 'East']
}
df_csv = pd.DataFrame(csv_data)

# Sample data for Excel format
excel_data = {
    'OrderID': [6, 7, 8],
    'Product': ['Product C', 'Product A', 'Product B'],
    'Sales': [210, 190, 220],
    'Region': ['North', 'East', 'West']
}
df_excel = pd.DataFrame(excel_data)

# Sample data for JSON format
json_data = [
    {'OrderID': 9, 'Product': 'Product A', 'Sales': 120, 'Region': 'South'},
    {'OrderID': 10, 'Product': 'Product B', 'Sales': 180, 'Region': 'East'}
]
# Convert JSON data to DataFrame
df_json = pd.DataFrame(json_data)
```

```python
# In a real scenario, replace these with pd.read_csv, pd.read_excel, pd.read_json
# Load CSV data
df_csv = pd.DataFrame(csv_data)

# Load Excel data
df_excel = pd.DataFrame(excel_data)

# Load JSON data
df_json = pd.DataFrame(json_data)
```

```python
# Check structure and content of CSV data
print("CSV data:")
print(df_csv.head())

# Check structure and content of Excel data
print("\nExcel data:")
print(df_excel.head())

# Check structure and content of JSON data
print("\nJSON data:")
print(df_json.head())
```

```python
# Merge into a single dataframe
df_merged = pd.concat([df_csv, df_excel, df_json], ignore_index=True)
```

```python
# Example: No transformation needed in this simplified example
```

```python
# Calculate total sales
total_sales = df_merged['Sales'].sum()

# Aggregate data by specific variables
sales_by_region = df_merged.groupby('Region')['Sales'].sum()

# Calculate average order value
average_order_value = df_merged.groupby('OrderID')['Sales'].mean()

# Product category distribution
product_category_distribution = df_merged['Product'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot of sales by region
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_region.index, y=sales_by_region.values)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie chart of product category distribution
plt.figure(figsize=(8, 8))
plt.pie(product_category_distribution.values, labels=product_category_distribution.index, autopct='%1.1f%%')
plt.title('Product Category Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
